```python
df1.pivot_table(index='subcategoria', aggfunc='count', values='nombre')
```

# **2. Procesamiento de los datos**

Normalizar toda la información de Museos, Salas de Cine y Bibliotecas
Populares, para crear una única tabla que contenga:
 
*   cod_localidad
*   id_provincia
*   id_departamento
*   categoría
*   provincia
*   localidad
*   nombre
*   domicilio
*   código postal
*   número de teléfono
*   mail
*   web


> ### A. Se normaliza el dataframe de museos:

```python
# Se eliminan las columnas innecesarias:
df1.drop(columns=['Observaciones', 
                  'categoria', 
                  'piso', 
                  'cod_area',
                  'Latitud', 
                  'Longitud', 
                  'TipoLatitudLongitud', 
                  'Info_adicional', 
                  'fuente',
                  'jurisdiccion',
                  'año_inauguracion',
                  'actualizacion'], 
                   inplace=True)
```

```python
# Se renombran las columnas restantes según indicaciones de normalización:
df1.rename(columns={'Cod_Loc': 'cod_localidad', 
                    'IdProvincia': 'id_provincia', 
                    'IdDepartamento': 'id_departamento',
                    'subcategoria': 'categoría', 
                    'direccion': 'domicilio', 
                    'CP': 'código postal',
                    'telefono': 'número de teléfono',
                    'Mail': 'mail',
                    'Web': 'web'}, inplace=True)
```

```python
# Se cambian los tipos de datos del dataframe resultante:
df1 = df1.astype('string')
```

```python
# Se cambian los valores null por el string "s/d":
df1.fillna('s/d', inplace=True)
```

```python
df1.head()
```

> ### B. Se normaliza el dataframe de cines:

```python
# Se eliminan las columnas innecesarias:
df2.drop(columns=['Observaciones', 
                  'Departamento',
                  'Piso', 
                  'cod_area',
                  'Latitud', 
                  'Longitud', 
                  'TipoLatitudLongitud', 
                  'Información adicional', 
                  'Fuente',
                  'tipo_gestion',
                  'Pantallas',
                  'Butacas',
                  'espacio_INCAA',
                  'año_actualizacion'], 
                   inplace=True)
```

```python
# Se renombran las columnas restantes según indicaciones de normalización:
df2.rename(columns={'Cod_Loc': 'cod_localidad', 
                    'IdProvincia': 'id_provincia', 
                    'IdDepartamento': 'id_departamento',
                    'Categoría': 'categoría', 
                    'Provincia': 'provincia',
                    'Localidad': 'localidad',
                    'Nombre': 'nombre',
                    'Dirección': 'domicilio', 
                    'CP': 'código postal',
                    'Teléfono': 'número de teléfono',
                    'Mail': 'mail',
                    'Web': 'web'}, inplace=True)
```

```python
# Se cambian los tipos de datos del dataframe resultante:
df2 = df2.astype('string')
```

```python
# Se cambian los valores null por el string "s/d":
df2.fillna('s/d', inplace=True)
```

```python
df2.head()
```

> ### C. Se normaliza el dataframe de bibliotecas:

```python
df3.dtypes
```

```python
# Se eliminan las columnas innecesarias:
df3.drop(columns=['Observacion', 
                  'Subcategoria',
                  'Departamento',
                  'Piso',
                  'Cod_tel',
                  'Latitud', 
                  'Longitud', 
                  'TipoLatitudLongitud', 
                  'Información adicional', 
                  'Fuente',
                  'Tipo_gestion',
                  'año_inicio',
                  'Año_actualizacion'], 
                   inplace=True)
```

```python
# Se renombran las columnas restantes según indicaciones de normalización:
df3.rename(columns={'Cod_Loc': 'cod_localidad', 
                    'IdProvincia': 'id_provincia', 
                    'IdDepartamento': 'id_departamento',
                    'Categoría': 'categoría', 
                    'Provincia': 'provincia',
                    'Localidad': 'localidad',
                    'Nombre': 'nombre',
                    'Domicilio': 'domicilio', 
                    'CP': 'código postal',
                    'Teléfono': 'número de teléfono',
                    'Mail': 'mail',
                    'Web': 'web'}, inplace=True)
```

```python
# Se cambian los tipos de datos del dataframe resultante:
df3 = df3.astype('string')
```

```python
# Se cambian los valores null por el string "s/d":
df3.fillna('s/d', inplace=True)
```

```python
df3.head(2)
```

> ### D. Se unen los tres dataframes en una única tabla:

```python
frames = [df1, df2, df3]

df = pd.concat(frames, ignore_index=True)
```

```python
df
```

```python
#Se exporta el dataframe a un nuevo archivo excel
df.to_excel('tabla única - museos_cines_bibliotecas.xlsx', 
              sheet_name='Tabla única')
```
